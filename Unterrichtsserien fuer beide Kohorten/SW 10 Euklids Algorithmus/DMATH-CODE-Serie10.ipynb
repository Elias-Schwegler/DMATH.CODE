```python
''' Bestimmt die Primfaktorzerlegung von n mit Probedivision'''

def primeFactor(n):
    factor = 2
    while factor <= n:
        if n % factor == 0:
            print(factor,end=" ")
            n = n // factor
        else:
            factor += 1
    print()
```

```python
''' Primfaktorzerlegung für die ersten 100 Zahlen '''

for i in range(2,101):
    print(str(i) + " = ",end=" ")
    primeFactor(i) 
```

```python
''' 11-te Mersenne Zahl ist nicht Prim '''

primeFactor(2**11 - 1)
    
```

```python
''' 24-stellige Zahl mit kleinen Primfaktoren '''    

primeFactor(484639526894037745950720)
```

```python
''' 24-stellige Zahl mit grossen Primfaktoren -> terminiert nicht mehr zeitnah '''
primeFactor(565765434324543216797351)
```
