**Simulation des Monty-Hall-Problem**

In der amerikanischen Spielshow *Let's make a deal* wurde folgendes Spiel angeboten: Sie haben die Wahl zwischen drei Türen. Hinter einer der Türen ist ein Auto, hinter den anderen sind Ziegen. Das Auto und die Ziegen sind vor der Show zufällig auf die Türen verteilt worden. Sie haben keine Information über die Position des Autos. Der Moderator weiss, was sich hinter den Türen befindet. Die Spielregeln lauten:
- Sie wählen zuerst eine Tür aus. Diese bleibt geschlossen.
- Der Moderator muss nun eine der beiden verbleibenden Türen öffnen. Hinter der von ihm geöffneten Tür muss sich eine Ziege befinden.
- Nachdem der Moderator eine Tür mit einer Ziege geöffnet hat, fragt er Sie, ob Sie bei Ihrer ersten Wahl bleiben oder zur letzten verbliebenen Tür wechseln möchten.

**Soll man wechsel?**

```python
import random

# Spielstrategie: nicht wechseln
numGames = 1000000
playerA_winCar = 0  # bleibt bei seiner ersten Wahl 
playerA_winGoat = 0
playerB_winCar = 0  # wechselt nach der ersten Wahl
playerB_winGoat = 0

gates = ['goat','goat','car']

for _ in range(numGames):
    random.shuffle(gates)
    choise = random.randint(0,2)
    if gates[choise] == 'car':
        playerA_winCar += 1     # bleibt beim Auto
        playerB_winGoat += 1    # wechselt zu der Ziege
    else:
        playerA_winGoat += 1    # bleibt bei der Ziege
        playerB_winCar +=1      # wechselt zum Auto, 
                                # weil andere Zeige ja gezeigt wird

print('Spielstrategie ohne Wechseln')
print('Auto gewonnen:', playerA_winCar/numGames)
print('Ziege gewonnen: ', playerA_winGoat/numGames,end='\n\n')

print('Spielstrategie mit Wechseln')
print('Auto gewonnen:', playerB_winCar/numGames)
print('Ziege gewonnen: ', playerB_winGoat/numGames)
```

<font size ="5">Die Koeffizienten des Polynoms $\textcolor{red}{\pmb{1}}\cdot x^6+\textcolor{red}{\pmb{1}}\cdot x^5+\textcolor{red}{\pmb{1}}\cdot x^4+\textcolor{red}{\pmb{1}}\cdot x^3+\textcolor{red}{\pmb{1}}\cdot x^2+\textcolor{red}{\pmb{1}}\cdot x^1+\textcolor{red}{\pmb{0}}\cdot x^0$ werden vom grössten Potenzterm $x^6$ zum kleinsten $x^0$ gespeichert.  </font>

```python
p = np.poly1d([1,1,1,1,1,1,0])
generatingFunction = p**3
generatingFunction
```

<font size ="5">Mit dem Attribut $\texttt{c}$ können die Koeffizienten des Polynoms mit dem Slice Operator $\texttt{[ ]}$ herausgelesen werden und das Attribut $\texttt{o}$ gibt den Grad des Polynoms zurück: </font>

```python
degree = generatingFunction.o 
degree
```

```python
# Koeffizienten a von x^9, x^8,...,x^1
sum(generatingFunction.c[degree-9:degree+1]) 
```
