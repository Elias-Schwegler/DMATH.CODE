Kann mit der Methode $\texttt{randint()}$ aus dem Modul $\texttt{random}$ ein fairer Würfel simuliert werden?

```python
import numpy as np
import random as rd
rd.randint(1,6)
```

Simulieren wir 5000000 Würfe und bestimmen die Häufigkeiten der Augenzahlen:

```python
Augenzahlen = np.array([0,0,0,0,0,0])
Anzahl_Spiele = 5000000
for _ in range(Anzahl_Spiele):
    Wurf = rd.randint(1,6)
    Augenzahlen[Wurf-1] += 1
np.round(Augenzahlen/Anzahl_Spiele,decimals=4)
```

Simulieren wir die zwei Spiele von Chevalier de Mere:

- Beim **Spiel_1** wird 4-mal ein Würfel geworfen. <br>
  Gewonnen wird das Spiel, wenn mindestens eine Sechs vorkommt.
- Beim **Spiel_2** werden 24-mal zwei Würfel geworfen. <br>
  Gewonnen wird das Spiel, wenn mindestens eine Doppelsechs vorkommt.

```python
Gewinne_Spiel_1 = 0
Gewinne_Spiel_2 = 0

Anzahl_Spiele = 500000
for _ in range(Anzahl_Spiele):
    Spielresultat = [rd.randint(1,6) for _ in range(4)]
    if 6 in Spielresultat:
        Gewinne_Spiel_1 += 1

for _ in range(Anzahl_Spiele):
    Spielresultat = [(rd.randint(1,6),rd.randint(1,6)) for _ in range(24)]
    if (6,6) in Spielresultat:
        Gewinne_Spiel_2 += 1

print(np.round(np.array([Gewinne_Spiel_1,Gewinne_Spiel_2])/Anzahl_Spiele,decimals=4))
```
