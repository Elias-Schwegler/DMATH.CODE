```python
import numpy as np

T = 1/12*np.array([[0,6,6],[3,6,3],[4,4,4]])
T.sum(axis=1)
np.matmul(T,T)
```

```python
startV = np.array([1,0,0])
invarianteV = startV
for _ in range(25):
    invarianteV =  np.matmul(invarianteV,T)
    print(np.round(invarianteV,4))
```
