```python
''' Systemmatrix '''
N = np.array([[0/5, 3/5, 0/5, 0/4, 2/7, 1/3],
              [3/5, 0/5, 1/5, 0/4, 2/7, 1/3] ,
              [1/5, 0/5, 0/5, 3/4, 1/7, 1/3],
              [0/5, 0/5, 2/5, 0/4, 1/7, 0/3],
              [0/5, 2/5, 2/5, 1/4, 0/7, 0/3],
              [1/5, 0/5, 0/5, 0/4, 1/7, 0/3]])

''' Startvektor '''
s_0 = 1/6*np.array([[1],[1],[1],[1],[1],[1]])

''' Matrix-Vektor-Iteration ''' 
s = 1 / np.sqrt((np.dot(s_0.T,s_0))) * s_0 # normieren
# Iterationen
anzIter = 0
for _ in range(500):
    anzIter +=1
    s_alt = s
    s_next = np.matmul(N,s)
    s = 1 / np.sqrt((np.dot(s_next.T,s_next))) * s_next # normieren
    ''' Abbruchbedingung '''
    if np.abs((s_alt-s)).sum() < 10**(-10): 
        break

print("Verteilung der Sympathiegehalte:",np.round(1/s.sum()*s*100,4),
      "\nAnzahl Iterationen:", anzIter,
      "\nSummentest:",(np.round(1/s.sum()*s*100)).sum())
```
